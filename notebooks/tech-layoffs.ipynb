### Cleaning up Data Type
Stock Reaction: this category seems to have two outcomes "positive" or "negative" so it makes the most sense to save space and change the values to a categorical

Layoff Size Category: Category being in the name kind of gave this one away, but it helps to have this as a categorical for modeling purposes

```python
df['stock_reaction'] = df['stock_reaction'].astype('category')
df['layoff_size_category'] = df['layoff_size_category'].astype('category')
```
